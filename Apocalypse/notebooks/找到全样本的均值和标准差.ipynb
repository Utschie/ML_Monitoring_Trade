```python
#采用分步计算法，先遍历一遍计算整体均值
mean_list = []
len_list = []
for i in tqdm(filelist):
    data = pd.read_csv(i)#读取文件
    data = data.drop(columns=['league','zhudui','kedui','companyname'])#去除非数字的列
    new_data = np.array(data)
    new_data = np.delete(new_data,(0,1), axis=-1)
    new_data = new_data[:,0:4]
    mean_list.append(new_data.mean(axis=0))
    len_list.append(new_data.shape[0])
                    
mean_list = np.array(mean_list).transpose((1,0))#转置
len_list = np.array(len_list)
all_mean = np.dot(mean_list,len_list)/np.sum(len_list)#由此获得总均值
print('已求得总平均值：'+str(all_mean))
var_sum = np.array([0.,0.,0.,0.])
for i in tqdm(filelist):
    data = pd.read_csv(i)#读取文件
    data = data.drop(columns=['league','zhudui','kedui','companyname'])#去除非数字的列
    new_data = np.array(data)
    new_data = np.delete(new_data,(0,1), axis=-1)
    new_data = new_data[:,0:4]
    var_sum += np.sum((new_data-all_mean)**2,axis=0)

all_std = (var_sum/(np.sum(len_list)-1))**0.5
print('已求得总标准差：'+str(all_std))
    
    
```
